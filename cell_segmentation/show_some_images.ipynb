```python
train_df['id_cell_type'] = train_df['id'] + '/' + train_df['cell_type']
```

```python
cell_type_id = {}
for idx, row in train_df.iterrows():
    sub_id = row.id
    cell_type = row.cell_type
    if sub_id not in cell_type_id.keys():
        cell_type_id[sub_id] = []
    cell_types = cell_type_id[sub_id]
    if cell_type not in cell_types:
        cell_types.append(cell_type)
    cell_type_id[sub_id] = cell_types
```
